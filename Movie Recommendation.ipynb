```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Pre-Processing


```python
#Load the Dataset
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
# Selecting relevant features
features = ['title', 'genres', 'keywords', 'tagline', 'director', 'cast'] 
```

```python
#Replacing missoing values with empty strings
for feature in features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining all 5 features 
combined_features = movies_data['title'] + ' ' + movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['director'] + ' ' + movies_data['cast']
print(combined_features)
```

```python
#Converting text data into numerical vectors
vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combined_features)
print(feature_vector)
```

Cosine Similarity


```python
#Getting the cosine similarity matrix
similarity= cosine_similarity(feature_vector)
print(similarity)
print(similarity.shape)
```
